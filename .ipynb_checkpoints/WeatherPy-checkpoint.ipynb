```python
import random
import requests as req
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from citipy import citipy
import time
import json
from api_keys import weather_api_key as wkey
```

## Generate Cities List

```python
citipy_cities = []
weather_cities = []
i = 0
# get 600 cities 
while i<600:
    lon = random.uniform(-180,180)
    lat = random.uniform(-90, 90)
    city = citipy.nearest_city(lat,lon)
    
    # made some change to citipy definition so I can get the acctural geolocation coordinates from citipy dictionary
    if city.city_name not in [x["city"] for x in citipy_cities]:
        citipy_cities.append({
            "city" : city.city_name,
            "country" : city.country_code,
            "longtitude" : city.longtitude,
            "latitude" : city.latitude
        })
        i+=1
citipy_df = pd.DataFrame(citipy_cities)
citipy_df
```

```python
# unit: standard, metric and imperial
# use geolocation coordinates to get weather info
def url_by_coordinates(lat, lon, unit):
    return f"http://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&units={unit}&appid={wkey}"

data = []
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# get start time
now = round(time.time())
total_time = 0
for index, row in citipy_df.iterrows():
    print(f"Processing city {row['city']} of {row['country']} at latitude {row['latitude']} and longtitude {row['longtitude']}")
    # get info with imperial units
    response = req.get(url_by_coordinates(row["latitude"], row["longtitude"], "imperial"))
    data.append(response.json())
    # processing 60 calls and wait
    if index % 60 == 59:
        new_time = round(time.time())
        interval =  new_time - now
        total_time = total_time + interval
        now = new_time
        print(f"{index + 1} cities have been processed, totally used {total_time} seconds. Last api calls used {interval} seconds.")
        print("-------------------")
        # openweathermap api limits to 60 calls per minute, to wait until 1 minute before next 60 calls
        if interval < 60:
            print(f"Wait {60 - interval + 5} seconds to continue....")
            time.sleep(60 - interval + 5)
# show processing done
print("All done.")
```

```python
print(len(data))
#print(json.dumps(data[0], indent=4, sort_keys=True))
w_list = []
for x in data:
    # as using api by geographic coordinates, some data without city & country value
    try:
        w_list.append({
            "city" : x["name"],
            "country" : x["sys"]["country"],
            "lat" : x["coord"]["lat"],
            "lon" : x["coord"]["lon"],
            "temp" : x["main"]["temp"],
            "temp_max" : x["main"]["temp_max"],
            "temp_min" : x["main"]["temp_min"],
            "humidity" : x["main"]["humidity"],
            "cloudness" : x["clouds"]["all"],
            "wind" : x["wind"]["speed"]
        })
    except:
        w_list.append({
            "city" : None,
            "country" : None,
            "lat" : x["coord"]["lat"],
            "lon" : x["coord"]["lon"],
            "temp" : x["main"]["temp"],
            "temp_max" : x["main"]["temp_max"],
            "temp_min" : x["main"]["temp_min"],
            "humidity" : x["main"]["humidity"],
            "cloudness" : x["clouds"]["all"],
            "wind" : x["wind"]["speed"]
        })
    
df = pd.DataFrame(w_list)
```

```python
# remove data without city name and duplicated data
df = df.dropna(how="any")
df = df.drop_duplicates(keep="first").reset_index(drop=True)
df["city"].count()
```
